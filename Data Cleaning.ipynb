# Cleaning Adult Dataset from Kaggle

```python
import pandas as pd
```

```python
#Read the data from the dataset using "PANDAS DATAFRAME"
df=pd.read_csv(r"C:\GUVI\D101D102D103\Datasets\adult.csv")
df
```

```python
# Removing Duplicate rows from the data set.
# Note the no of rows in the above line and the no of rows in the below line
df.drop_duplicates()
```

```python
# Assigning variable df(DataFrame) to the function
df=df.drop_duplicates()
df
```

```python
# Removing the column with no meaning or insights.
df=df.drop(columns='fnlwgt')
df

# Note: Column Capital-loss has only 0's in it. So we can drop the column in the next line of code.
```

```python
#Removing the column Capital-loss in the code
df=df.drop(columns="capital-loss")
df
```

```python
#Print the dataframe and reassure the table doesn't contain both the removed columns
df
```

```python
# Noticing "?" value in the dataset 
# Removing the "?" value using str.strip() function
df['occupation']=df['occupation'].str.strip("?")
df
```

```python
# Using the above method to remove it in "education column"
df['education']=df['education'].str.strip("?")
df
```

```python
#Repeating the process in other columns
df['workclass']=df['workclass'].str.strip("?")
df['native-country']=df['native-country'].str.strip("?")
df
```

```python
#Verify the dataset for presence of "?"
df
```

```python
#removing duplicate values
df=df.drop_duplicates()
df
```

```python
#Replacing "Nan" values with space/nothing
df=df.replace("Nan","")
df
```

```python
df["occupation"].unique()
```

```python
#Replacing values with right words

df["occupation"]=df["occupation"].replace("Protective-serv","Protective-service")
df["occupation"]=df["occupation"].replace("Adm-clerical","Admin-clerical")
df["occupation"]=df["occupation"].replace("Priv-house-serv","Private-house-service")
df["occupation"]=df["occupation"].replace("Exec-managerial","Executive-managerial")
df["occupation"]=df["occupation"].replace("Machine-op-inspct","Machine-opertor-inspctor")
df
```

```python
# Removing duplicates and resetting the indices
df=df.drop_duplicates()
df=df.reset_index(drop=True)
df
```

```python
df.to_csv("Adult_cleaned_dataset.csv")
df.to_excel("Adult_cleaned_dataset.xlsx")
```
